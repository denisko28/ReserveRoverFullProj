# Business(Places) dataset

```python
data_file = open("yelp_dataset/yelp_academic_dataset_business.json")
data = []
for line in data_file:
  data.append(json.loads(line))
places_df = pd.DataFrame(data)
data_file.close()
```

```python
places_df.head()
```

```python
print(places_df.info())
```

```python
places_df = places_df[places_df['address'].notna() & (places_df['address'] != '')]
```

```python
mapped_places_df = pd.DataFrame()

business_id_placeId_map = {row['business_id']: idx+100 for idx, row in places_df.iterrows()}
mapped_places_df['id'] = list(business_id_placeId_map.values())

mapped_places_df['manager_id'] = ['some_manager_id'] * len(places_df)
mapped_places_df['city_id'] = np.random.randint(1, 3, len(places_df))
mapped_places_df['main_image_url'] = ['https://ibb.co/87TPzVV'] * len(places_df)
mapped_places_df['title'] = places_df['name'].str.replace('"', '')
mapped_places_df['opens_at'] = pd.to_datetime('08:00:00').time()
mapped_places_df['closes_at'] = pd.to_datetime('18:00:00').time()
mapped_places_df['avg_mark'] = places_df['stars']
mapped_places_df['Popularity'] = np.random.randint(0, 300, len(places_df))
mapped_places_df['avg_bill'] = np.random.randint(300, 5000, len(places_df))
mapped_places_df['address'] = places_df['address'].str.replace('"', '')
mapped_places_df['ImagesCount'] = 1
mapped_places_df['SubmissionDateTime'] = datetime.now()
mapped_places_df['public_date'] = datetime.now().date()
mapped_places_df['moderation_status'] = 2

mapped_places_df['avg_mark'] = mapped_places_df['avg_mark'].astype(float).round(1)
mapped_places_df['Popularity'] = mapped_places_df['Popularity'].astype(int)
mapped_places_df['avg_bill'] = mapped_places_df['avg_bill'].astype(float).round(2)
mapped_places_df['ImagesCount'] = mapped_places_df['ImagesCount'].astype(int)
mapped_places_df['moderation_status'] = mapped_places_df['moderation_status'].astype(int)
```

```python
mapped_places_df.head()
```

```python
mapped_places_df.info()
```

```python
mapped_places_df.to_csv(r'places.csv', sep='\t', encoding='utf-8', index=False)
```

# Reviews dataset

```python
data_file = open("yelp_dataset/yelp_academic_dataset_review.json")
data = []
for line in data_file:
  data.append(json.loads(line))
```

```python
reviwes_df = pd.DataFrame(data)
data_file.close()
```

```python
reviwes_df.head()
```

```python
reviwes_df.info()
```

Фільтруємо дані в reviwes_df, щоб прибрати відгуки, для яких не існує зв'язаних користувачів

```python
mapped_users_df = pd.read_csv("users.csv")
```

```python
reviwes_df = reviwes_df[reviwes_df['user_id'].isin(mapped_users_df.iloc[:, 0])]
```

Фільтруємо дані в reviwes_df, щоб прибрати відгуки, для яких не існує зв'язаних закладів

```python
reviwes_df = reviwes_df[reviwes_df['business_id'].isin(list(business_id_placeId_map.keys()))]
```

```python
reviwes_df.info()
```

Прибираємо стовпці що не містять корисних даних

```python
reviwes_df = reviwes_df.drop(columns=['review_id', 'useful', 'funny', 'cool'])
```

Перевіряємо дані на наявність порожніх значень

```python
print(reviwes_df.isnull().sum())
```

Перевіряємо чи немає відгуків з відсутніми оцінками

```python
(reviwes_df['stars'] == 0).sum()
```

```python
reviwes_df.head()
```

```python
mapped_reviwes_df = pd.DataFrame()

mapped_reviwes_df['place_id'] = reviwes_df['business_id'].map(business_id_placeId_map)
mapped_reviwes_df['author_id'] = reviwes_df['user_id']
mapped_reviwes_df['creation_date'] = pd.to_datetime(reviwes_df['date']).dt.date
mapped_reviwes_df['mark'] = reviwes_df['stars'].astype(int)
mapped_reviwes_df['comment'] = reviwes_df['text']
```

```python
mapped_reviwes_df.head()
```

```python
mapped_reviwes_df['comment'] = mapped_reviwes_df['comment'].str.replace('"', "'")
```

```python
mapped_reviwes_df.replace('', np.nan, inplace=True)
```

```python
mapped_reviwes_df.isna().sum()
```

```python
print(mapped_reviwes_df.describe())
```

```python
print(f"Кількість унікальних авторів відгуків: {mapped_reviwes_df['author_id'].nunique()}")
print(f"Кількість унікальних закладів з відгуками: {mapped_reviwes_df['place_id'].nunique()}")
```

```python
mapped_reviwes_df = mapped_reviwes_df.head(1500000)
```

```python
mapped_reviwes_df.to_csv(r'reviews.csv', sep='|', encoding='utf-8', index=False)
```
